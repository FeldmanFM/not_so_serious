```python

with open('./group_prefix', 'r') as f:
    lGroupPrefix = f.readlines()
    lGroupPrefix = [i.replace('\n','') for i in lGroupPrefix]
f.close()


with open('./group_prefix_via_types2', 'r') as f:
    #print(f.readlines())
    lGroupPrefixTypes = f.readlines()
    lGroupPrefixTypes = [i.replace('\n','') for i in lGroupPrefixTypes]
    lGroupPrefixTypes = [i.split(',') for i in lGroupPrefixTypes]
f.close()


with open('./z1', 'r') as f:
    lAZ1 = f.readlines()
    lAZ1 = [i.replace('\n','') for i in lAZ1]
f.close()


with open('./z2', 'r') as f:
    lAZ2 = f.readlines()
    lAZ2 = [i.replace('\n','') for i in lAZ2]
f.close()
    
with open('./z3', 'r') as f:
    lAZ3 = f.readlines()
    lAZ3 = [i.replace('\n','') for i in lAZ3]
f.close()

with open('./very_new_cn_all', 'r') as f:
    lCN = f.readlines()
    lCN = [i.replace('\n','') for i in lCN]
    lCN = [i.split(',') for i in lCN]
f.close()

with open('./very_new_cn_migr', 'r') as f:
    lMigr = f.readlines()
    lMigr = [i.replace('\n','') for i in lMigr]
    lMigr = [i.split(',') for i in lMigr]
    lMigr = [i[1] for i in lMigr]
f.close()
```

```python
config = configparser.ConfigParser()
for i in sorted(lGroupPrefixTypes):
    #print(i)
    gp = i[0]
    gt = i[1]
    config[gp] = {}
    config[gp]['prefix'] = gp
    config[gp]['type'] = gt
    if gt == '6':
        config[gp]['migr'] = 'yes'
    else:
        config[gp]['migr'] = 'no'
with open('./conf_prefix_groups.ini', 'w') as configfile:
    config.write(configfile)
```

```python
config = configparser.ConfigParser()
for i in sorted(lCN, key = lambda x:  x[1]):
    #print(i)
    cn = i[0]
    cl = i[1]
    config[cl] = {}
    config[cl]['name'] = cn
    config[cl]['label'] = cl
    if cl in lMigr:
        config[cl]['migr'] = 'yes'
    else:
        config[cl]['migr'] = 'no'
    if cl in lAZ1:
        config[cl]['az'] = '1'
    elif cl in lAZ2:
        config[cl]['az'] = '2'
    elif cl in lAZ3:
        config[cl]['az'] = '3'    
with open('./conf_cn.ini', 'w') as configfile:
    config.write(configfile)
```
